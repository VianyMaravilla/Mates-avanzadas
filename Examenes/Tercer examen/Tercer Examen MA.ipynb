**Poblema 2:**

![Ejercicio 2](./imagen2.png "Ejercicio 2")

Construya las primeras 15 iteraciones del método del descenso pronunciado
(Steepest Descent Method), aplicado al problema anterior, considerando
como punto inicial el punto (0, 3). Presente los parámetros calculados en
cada iteración

```python
import numpy as np

def grad(x):
    grad = np.zeros(2)
    grad[0] = 4 * (x[0] - 2) ** 3 + 2 * (x[0] - 2 * x[1])
    grad[1] = -4 * (x[0] - 2 * x[1])
    return grad

def f(x):
    # funcion a minimizar
    return (x[0] - 2) ** 4 + (x[0] - 2 * x[1]) ** 2

x = np.array([0, 3]) # punto inicial
alpha = 0.01 # tamaño de paso
tolerance = 1e-6 # tolerancia para la convergencia
max_iter = 20 # numero maximo de iteraciones

# Steepest Descent 
for i in range(max_iter):
    grad_x = grad(x)
    x_new = x - alpha * grad_x
    if np.linalg.norm(x_new - x) < tolerance:
        break
    x = x_new
    print("Iteracion número: {}, x = {}, f(x) = {}".format(i, x, f(x)))
```

**Problema 3:**

Construya las primeras 15 iteraciones del méetodo de Newton aplicado al
problema del ejercicio 2, considerando como punto inicial (0, 3). Presente
los parámetros calculados en cada iteración

```python
def grad(x):
    x1, x2 = x[0], x[1]
    return np.array([4*(x1-2)**3 + 2*(x1-2*x2), 2*(x1-2*x2)*(-2)])

def hessian(x):
    x1, x2 = x[0], x[1]
    return np.array([[12*(x1-2)**2 + 2, -4], [-4, 2]])

x_k = np.array([0, 3])

for i in range(15):
    print(f"Iteration {i+1}: x_k = {x_k}")
    grad_k = grad(x_k)
    hessian_k = hessian(x_k)
    x_k = x_k - np.linalg.inv(hessian_k).dot(grad_k)
```
